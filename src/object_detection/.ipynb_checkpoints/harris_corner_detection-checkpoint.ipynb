```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_folder = "../../resource/"
```

```python
show_gray= lambda x:plt.imshow(x, cmap='gray')
```

```python
flat_chess_img = cv2.imread(f"{data_folder}/flat_chessboard.png")
gray_flat_chess = np.float32(cv2.cvtColor(flat_chess_img, cv2.COLOR_BGR2GRAY))
show_gray(gray_flat_chess)
```

```python
real_chess_img = cv2.imread(f"{data_folder}/real_chessboard.jpg")
gray_real_chess = np.float32(cv2.cvtColor(real_chess_img, cv2.COLOR_BGR2GRAY))
show_gray(gray_real_chess)
```

```python
dst = cv2.cornerHarris(src=gray_flat_chess,
                      blockSize=2,
                      ksize=3,
                      k=0.04)
show_gray(dst)
```

in the below image after performaing corner haris
we will get white dots on the corner, so where ever 
there is a white pixel inside result of cornerHarris 
we will add red color to the same pixel in real image

```python
dst = cv2.dilate(dst,None)
show_gray(dst)
# cv2.getStructuringElement(shape=3, ksize=(3,3))
```

```python
flat_chess_img[dst > 0.01*dst.max()] = [255,0,0]
show_gray(flat_chess_img)
```

```python
def get_corner(img):
    img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    dst = cv2.cornerHarris(src=img_grey,
                          blockSize=2,
                          ksize=3,
                          k=0.04)
    dst = cv2.dilate(dst, None)
    img[dst >0.01*dst.max()] = [255,0,0]
```
