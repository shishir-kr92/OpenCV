```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_folder = "../../resource/"
```

```python
flat_chess_img = cv2.imread(f"{data_folder}/flat_chessboard.png")
real_chess_img = cv2.imread(f"{data_folder}/real_chessboard.jpg")

flat_chess_img = cv2.cvtColor(flat_chess_img, cv2.COLOR_BGR2RGB)
real_chess_img = cv2.cvtColor(real_chess_img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_chess_img)
```

```python
def get_corners(img):
    img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    corners = cv2.goodFeaturesToTrack(img_gray,
                                 100,
                                 0.01,
                                 10)
    for i in np.int0(corners):
        x, y = i.ravel()
        cv2.circle(img, 
                  center=(x,y),
                  radius=4,
                  color=(255,0,0),
                  thickness=-1)
    return corners
```

```python
corners = get_corners(flat_chess_img)
plt.imshow(flat_chess_img)
```

```python
corners = get_corners(real_chess_img)
plt.imshow(real_chess_img)
```
