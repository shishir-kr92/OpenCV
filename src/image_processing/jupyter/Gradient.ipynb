```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## This 
```

```python
def display_img(img):
    fig = plt.figure(figsize=(8,6))
#     fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
sudoko_img = cv2.imread('../../../resource/sudoku.jpg', 0)
display_img(sudoko_img)
```

```python
sobel_x = cv2.Sobel(sudoko_img, cv2.CV_64F,1,0, ksize=5)
display_img(sobel_x)
```

```python
sobel_y = cv2.Sobel(sudoko_img, cv2.CV_64F,0,1, ksize=5)
display_img(sobel_y)
```

```python
laplace_img = cv2.Laplacian(sudoko_img, cv2.CV_64F)
laplace_img.min()
laplace_img.max()
display_img(laplace_img)
```

```python
blended = cv2.addWeighted(src1=laplace_img, 
                          alpha=0.5,
                          src2=sobel_y,
                          beta=0.5,
                          gamma=100)
display_img(blended)
```

```python
ret, th1 = cv2.threshold(sudoko_img, 100, 255, cv2.THRESH_BINARY)
display_img(th1)
```

```python
kernel = np.ones((4,4), dtype=np.uint8)
gradient = cv2.morphologyEx(blended, cv2.MORPH_GRADIENT, kernel)
display_img(gradient)
```
