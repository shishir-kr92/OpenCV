```python
## To Blend Image
# new_pixal = alpha * pixel_1 + beta * pixel_2 + gamma
# where alpha, beta and gamma are some constant
```

```python
import cv2
```

```python
path = "../../../resource"
img1 = cv2.imread(f"{path}/dog_backpack.png")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread(f"{path}/watermark_no_copy.png")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img1.shape
```

```python
img2.shape
```

```python
img1 = cv2.resize(img1, dsize=(1200, 1200))
img2 = cv2.resize(img2, dsize=(1200, 1200))
```

```python
blended = cv2.addWeighted(src1=img1, alpha=0.65, src2=img2, beta=0.35, gamma=0)
```

```python
plt.imshow(blended)
```
