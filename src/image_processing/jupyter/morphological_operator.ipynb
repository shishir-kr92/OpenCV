```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

def load_img():
    blank_img = np.zeros(shape=(600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='ABCD',org=(60, 500), fontFace=font,fontScale=5, color = (255, 0,0), thickness=30 )
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_img()
display_img(img)
```

```python
## erosion will erode the bondries of foreground object
kernel = np.ones((5,5))
```

```python
dst = cv2.erode(img, kernel, iterations=5)
display_img(dst)
```

```python
## create noisey image
img = load_img()
img
```

```python
white_noise = np.random.randint(low=0, high=2, size=(600,600)) * 255
white_noise
```

```python
noisy_img = img + white_noise
display_img(noisy_img)
```

```python
dst = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
display_img(dst)
```
