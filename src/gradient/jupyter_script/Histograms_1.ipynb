```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
file_path="../../../resource"
dark_horse = cv2.imread(f"{file_path}/horse.jpg")
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB)

rainbow = cv2.imread(f"{file_path}/rainbow.jpg")
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

blue_brick = cv2.imread(f"{file_path}/bricks.jpg")
show_brick = cv2.cvtColor(blue_brick, cv2.COLOR_BGR2RGB)

plt.imshow(show_brick)
show_brick.shape
```

```python
histr = cv2.calcHist([blue_brick],
                    channels=[0],
                    mask=None,
                    histSize=[256],
                    ranges=[0,256])
plt.plot(histr)
```

```python
img= show_brick
color = ('b', 'g', 'r')
```

```python
for i,col in enumerate(color):
    histr = cv2.calcHist([blue_brick],
                         [i], 
                         mask=None,
                         histSize=[256],
                         ranges=[0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,256])
plt.title("Color HistoGram")
```

```python
for i,col in enumerate(color):
    histr = cv2.calcHist([rainbow],
                         [i], 
                         mask=None,
                         histSize=[256],
                         ranges=[0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,256])
plt.title("Color HistoGram")
```

```python
for i,col in enumerate(color):
    histr = cv2.calcHist([dark_horse],
                         [i], 
                         mask=None,
                         histSize=[256],
                         ranges=[0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,50])
    plt.ylim([0,50000])
plt.title("Color HistoGram")
```
